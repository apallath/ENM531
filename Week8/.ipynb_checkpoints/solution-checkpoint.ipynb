```python
# convert label for 2, 5 class: 2 --> 0, 5 --> 1 and 7 --> 2
def convert_label_1(labels):
    N = labels.shape[0]
    for i in range(N):
        if labels[i] == 2:
            labels[i] = 0
        elif labels[i] == 5:
            labels[i] = 1
        else:
            labels[i] = 2
    return labels
            
# invert label for 0, 1, 2 class: 0 --> 2, 1 --> 5 and 2 --> 7
def invert_label_1(labels):
    N = labels.shape[0]
    for i in range(N):
        if labels[i] == 0:
            labels[i] = 2
        elif labels[i] == 1:
            labels[i] = 5
        else:
            labels[i] = 7
    return labels
            
```

```python
def one_hot(y):
    N, C = y.shape[0],y.max(0)+1
    Y = np.zeros((N,C))
    for i in range(0, N):
        idx = y[i]
        Y[i,idx] = 1
    return Y

def deone_hot(Y):
    N = Y.shape[0]
    y = np.zeros(N)
    for i in range(0, N):
        y[i] = Y[i,:].argmax()
    return y
```

```python
# Define the forward function of the CNN

class Forward_CNN(torch.nn.Module): 
    def __init__(self):
        super(Forward_CNN, self).__init__()
        self.conv1 = torch.nn.Conv2d(1, 16, 3, 1, 1)
        self.conv2 = torch.nn.Conv2d(16, 16, 3, 1, 1)
        self.fc1 = torch.nn.Linear(784, 64)
        self.fc2 = torch.nn.Linear(64, 3)

    def forward(self, x):   # M x 1 x 28 x 28 
        x = self.conv1(x)
        x = F.relu(x)
        x = F.max_pool2d(x, 2)
        x = self.conv2(x)
        x = F.relu(x)
        x = F.max_pool2d(x, 2)
        x = torch.flatten(x, 1)
        x = self.fc1(x)
        x = F.relu(x)
        x = self.fc2(x)
        output = F.softmax(x, dim=1)
        return output
```

```python
class CNN_Classification:
    # Initialize the class
    def __init__(self, X, Y):  
      
        # Check if there is a GPU available
        if torch.cuda.is_available() == True:
            self.dtype_double = torch.cuda.FloatTensor
            self.forward_cnn = Forward_CNN().cuda()
        else:
            self.dtype_double = torch.FloatTensor
            self.forward_cnn = Forward_CNN()
           
        self.jitter = 1e-6
        
        X = np.expand_dims(X, axis=1)
        
        self.Xmean, self.Xstd = X.mean(0), X.std(0)
        X = (X - self.Xmean) / (self.Xstd + self.jitter)
        

        # Define PyTorch dataset
        self.X = torch.from_numpy(X).type(self.dtype_double) # num_images x num_channels x num_pixels_x x num_pixels_y
        self.Y = torch.from_numpy(Y).type(self.dtype_double) # num_images x 3

        # Define the optimizer
        self.optimizer = torch.optim.Adam(list(self.forward_cnn.parameters()), lr=1e-3, betas=(0.9, 0.999))
     
        self.loss_train = []
    
    def forward_pass(self, X):  
        Y = self.forward_cnn(X)  
        return Y 

    
    def compute_loss(self, X_batch, Y_batch):
        Y_pred = self.forward_pass(X_batch)
        loss = Y_batch*torch.log(Y_pred + self.jitter)            
        return - torch.sum(loss)
    
    
    # Fetches a mini-batch of data
    def fetch_minibatch(self,X, Y, N_batch):
        N = X.shape[0]
        idx = np.random.choice(N, N_batch, replace=False)
        X_batch = X[idx,:]
        Y_batch = Y[idx,:]        
        return X_batch, Y_batch
    
    
    def train(self, nIter = 10000, batch_size = 100):   

        start_time = timeit.default_timer()            
        for it in torch.arange(nIter):
            X_batch, Y_batch = self.fetch_minibatch(self.X, self.Y, batch_size)
            
            self.optimizer.zero_grad()
            loss_train = self.compute_loss(X_batch, Y_batch)   

            loss_train.backward()            
            self.optimizer.step()
            
            # Print
            if it % 100 == 0:
                elapsed = timeit.default_timer() - start_time
                print('It: %d, Loss: %.3e, Time: %.2f' % 
                      (it, loss_train, elapsed))
                start_time = timeit.default_timer()
    
            self.loss_train.append(loss_train)
        
        
    def predict(self, X_star): 
        X_star = np.expand_dims(X_star, axis=1)
        X_star = (X_star - self.Xmean) / (self.Xstd + self.jitter)           
        X_star = torch.from_numpy(X_star).type(self.dtype_double)
        Y_star = self.forward_pass(X_star)
        Y_star = Y_star.cpu().data.numpy()
        return Y_star

```

```python
test_data_1 = test_data[test_idx_1,:,:]
train_data_1 = train_data[train_idx_1,:,:]
test_labels_1 = test_labels[test_idx_1]
train_labels_1 = train_labels[train_idx_1]

train_labels_1 = convert_label_1(train_labels_1)
test_labels_1 = convert_label_1(test_labels_1)

N_train = train_data_1.shape[0]
N_test = test_data_1.shape[0]

train_data_final = train_data_1
test_data_final = test_data_1
train_labels_final = one_hot(train_labels_1)
test_labels_final = one_hot(test_labels_1)
```

```python
model  = CNN_Classification(train_data_final, train_labels_final)

# Training
model.train(nIter = 500, batch_size = 128)

# Predictions
pred_labels = model.predict(test_data_final)

predicted_label = deone_hot(pred_labels)
test_labels_final = deone_hot(test_labels_final)
```
